```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_score = [40, 45, 50, 55, 60, 65, 75, 85, 90]
 
import matplotlib.pyplot as plt
colors = ["b" if score>57 else "r" for score in exam_score]
plt.scatter(study_hours,exam_score,c = colors, s = 30)
plt.title("study_hours  v/s  exam_scores")
plt.xlabel("study_hours")
plt.ylabel("exam_scores")
plt.grid(True,linestyle = "--",alpha = 0.3)
```

```python
plt.scatter(study_hours,exam_score,c = exam_score,cmap = "YlOrBr")
plt.colorbar(label = 'exam_score')
plt.title("study_hours  v/s  exam_scores")
plt.xlabel("study_hours")
plt.ylabel("exam_scores")
for i in range(len(study_hours)):
    plt.text(study_hours[i]+0.2,exam_scores[i],f"s{i}",fontstyle = "italic")
plt.grid(True,linestyle = "--",alpha = 0.3)
```

```python
social_score = [200,340,280,540,386,487,197,498,500]
plt.scatter(study_hours,exam_score,c = social_score,cmap = "YlOrBr")
plt.colorbar(label = 'social score')
plt.title("study_hours  v/s  exam_scores")
plt.xlabel("study_hours")
plt.ylabel("exam_scores")
plt.grid(True,linestyle = "--",alpha = 0.3)
```

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours,class_a_scores,label = "class A",c = "b")
plt.scatter(class_b_hours,class_b_scores,label = "class B",c = "r")

for i in range(len(class_a_hours)):
    plt.text(class_a_hours[i]+0.2,class_a_scores[i],f"A{i}",fontstyle = "italic")

for i in range(len(class_b_hours)):
    plt.text(class_b_hours[i]+0.2,class_b_scores[i],f"B{i}",fontstyle = "italic")

plt.title("class A  v/s  class B")
plt.xlabel("hours")
plt.ylabel("scores")
plt.legend()
plt.grid(True,linestyle = "--",alpha = 0.3)

```
