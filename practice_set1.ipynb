```python
import matplotlib.pyplot as plt
```

```python
list1 = [1,2,3]
list2 = [4,5,6]
plt.plot(list1,list2)
plt.xlabel("list1")
plt.ylabel("list2")
plt.title("just an example")
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kohli)
plt.xlabel("years")
plt.ylabel("runs made")
plt.title("runs made by kohli year wise")
```

```python
plt.plot(years,kohli,"or")
plt.xlabel("years")
plt.ylabel("runs made")
plt.title("runs made by kohli year wise")
```

```python
plt.plot(years,kohli,linestyle = "--",marker = "^",color = "b",linewidth = 3)
plt.xlabel("years")
plt.ylabel("runs made")
plt.title("runs made by kohli year wise")
```

```python
plt.plot(years,kohli,label = "virat kohli")
plt.plot(years,sehwag,label = "sehwag")
plt.xlabel("years")
plt.ylabel("runs made")
plt.title("runs made kohli and sehwag year wise")
plt.legend()
```

```python
plt.plot(years,kohli,"o-r",label = "virat kohli")
plt.plot(years,sehwag,"o-b",label = "sehwag")
plt.xlabel("years")
plt.ylabel("runs made")
plt.title("runs made kohli and sehwag year wise")
plt.legend()
```

```python
plt.plot(years,kohli,"o-r",label = "virat kohli")
plt.plot(years,sehwag,"o-b",label = "sehwag")
plt.xlabel("years")
plt.ylabel("runs made")
plt.title("runs made kohli and sehwag year wise")
plt.legend()
plt.grid(True)
```

```python
with plt.xkcd():
    plt.plot(years,kohli,"o-r",label = "virat kohli")
    plt.plot(years,sehwag,"o-b",label = "sehwag")
    plt.xlabel("years")
    plt.ylabel("runs made")
    plt.title("runs made kohli and sehwag year wise")
    plt.legend()
    plt.grid(True)
```

```python
plt.plot(years,kohli,"o-r",label = "virat kohli")
plt.plot(years,sehwag,"o-b",label = "sehwag")
plt.xlabel("years")
plt.ylabel("runs made")
plt.title("runs made kohli and sehwag year wise")
for i in range(len(years)):
    plt.text(years[i],kohli[i]-100,str(kohli[i]))

for i in range(len(years)):
    plt.text(years[i],sehwag[i]+50,str(sehwag[i]))
plt.legend()
plt.grid(True)
```
