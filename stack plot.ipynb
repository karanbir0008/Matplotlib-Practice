```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# Sample data: hours spent per activity each day
sleep = [7, 8, 6, 7, 7, 9, 8]
work  = [8, 9, 8, 8, 7, 0, 0]
exercise = [1, 1, 1, 1, 1, 2, 2]
leisure  = [3, 2, 4, 3, 5, 6, 7]
chores   = [2, 2, 2, 2, 2, 3, 3]
```

```python
plt.stackplot(days,sleep,work,exercise,leisure,chores,labels = ["sleep","work","exercise","leisure","chores"],alpha = 0.4)
plt.title("weekly schedule")
plt.xlabel("days")
plt.ylabel("activities")
plt.grid(True)
plt.legend(loc = "lower left")
```

```python
quarters = ["Q1", "Q2", "Q3", "Q4"]

product_a = [50, 70, 90, 85]
product_b = [40, 60, 55, 75]
product_c = [20, 30, 35, 45]

plt.figure(figsize=(4,4))
plt.stackplot(quarters,product_a,product_b,product_c,labels = ["product_a","product_b","product_c"],alpha = 0.4)
plt.xlabel("quarters")
plt.ylabel("products")
plt.title("sales of different products over time")
plt.grid()
```
