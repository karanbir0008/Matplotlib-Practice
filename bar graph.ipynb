```python
import matplotlib.pyplot as plt  
products = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Mangoes']
sales = [150, 200, 180, 90, 120]
```

```python


plt.bar(products,sales,width = 0.25,color = "b",edgecolor = "r")

plt.xlabel("fruits")
plt.ylabel("sales")
plt.title("sales of different fruits")
for i in range(len(products)):
    plt.text(i,sales[i]+2,str(sales[i]),color = "k",ha = "center")


```

```python
products = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Mangoes']
july_sales = [130, 180, 160, 100, 140]
august_sales = [150, 200, 180, 90, 120]

import numpy as np
x = np.arange(len(products))
bar_width = 0.3

plt.bar(x-bar_width/2,july_sales,label = "july month",width = 0.3)
plt.bar(x+bar_width/2,august_sales,label = "august month",width =0.3)
plt.xlabel("fruits")
plt.ylabel("sales")
plt.title("sales in july vs in august")


plt.xticks(x,products)
plt.tight_layout()
plt.legend()
```

```python
import numpy as np
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))  # index positions
width = 0.25
```

```python
plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
plt.show()
```

```python
years = [2018, 2019, 2020, 2021]

kohli  = [1200, 1300, 1250, 1400]
sachin = [1500, 1600, 1550, 1650]
sehwag = [1000, 1100, 950,  1200]
dhoni  = [800,  850,  900,  950]

x = np.arange(len(years))
width = 0.1
plt.bar(x-(0.1)*3/2,kohli,label = "kohli",width = width)
plt.bar(x-0.1/2,sachin,label = "sachin",width = width)
plt.bar(x+0.1/2,sehwag,label = "sehwag",width = width)
plt.bar(x+(0.1)*3/2,dhoni,label = "dhoni",width = width)

for i in range(len(years)):
    plt.text(x[i]-(0.1)*3/2,kohli[i]+30,str(kohli[i]),ha = "center")

for i in range(len(years)):
    plt.text(x[i]-0.1/2,sachin[i]+30,str(sachin[i]),ha = "center")

for i in range(len(years)):
    plt.text(x[i]+0.1/2,sehwag[i]+30,str(sehwag[i]),ha = "center")

for i in range(len(years)):
    plt.text(x[i]+(0.1)*3/2,dhoni[i]+30,str(dhoni[i]),ha = "center")

plt.xlabel("years")
plt.ylabel("runs")
plt.title("comparison")
plt.xticks(x,years)
plt.legend()
```

```python


y = np.arange(len(years))
height= 0.1
plt.barh(y-(0.1)*3/2,kohli,label = "kohli",height = height)
plt.barh(y-0.1/2,sachin,label = "sachin",height = height)
plt.barh(y+0.1/2,sehwag,label = "sehwag",height = height)
plt.barh(y+(0.1)*3/2,dhoni,label = "dhoni",height = height)

plt.xlabel("runs")
plt.ylabel("years")
plt.title("comparison")
plt.yticks(y,years)
plt.legend()
```
